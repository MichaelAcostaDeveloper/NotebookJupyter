```python
import matplotlib.pyplot as plt
import numpy as np
```

# **FUNCIÓN LINEAL**
---
La función lineal en su forma pendiente y ordenada en el origen es: 


> $y=mx+b$



Donde:
*    **m** es la pendiente de la recta, osea la inclinación.
*   **b** es el corte que tiene la recta con el eje y o eje de ordenadas.

Además es importante aclarar que el dominio de esta función son todos los números reales ℛ, osea no existe restricción alguna para los valores de x (la variable independiente).

      

```python
def f(m,x,b):
  return m*x+b
```

```python
#El número de puntos que se quiere analizar y crear
resolucion=int(input('Ingrese el valor de la resolucion de la pantalla: '))
m = float(input('Ingrese un valor de pendiente de la recta: '))
b=float(input('Ingres un valor de la interseccion con el eje y: '))
x_1=float(input('Ingrese el limite inferior para el valor de x: '))
x_2=float(input('Ingrese el limite superior para el valor de x: '))

#El intervalo de valores de x
x=np.linspace(x_1,x_2, num=resolucion) 
y=f(m,x,b)

#Este codigo permite dibujar 
fig, ax=plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0,color='r')     #para crear una linea horizontal
ax.axvline(x=0,color='r')     #para crear una linea vertical
```

# **FUNCIÓN CUADRÁTICA EN FORMA GENERAL**
---
La función cuadrática en su forma general es: 

> $y=f(x)=ax^{2} + bx + c$

Donde:
*    **a, b, c** son constantes que deben ser diferente de cero.

El dominio de esta función son todos los números reales ℛ, osea no existe restricción alguna para los valores de x (la variable independiente).

```python
def f(a,b,c,x): 
  return a*x**2+b*x+c

#Creacion de valores para graficar
a=float(input('Ingrese el valor del coeficiente a: '))
b=float(input('Ingrese el valor del coeficiente b: '))
c=float(input('Ingrese el valor del coeficiente c: '))

#Creacion de puntos y limites para la variable independiente x
resolucion=int(input('Ingrese el numero de puntos: '))
x_1=float(input('Ingrese el limite inferior para el valor de x: '))
x_2=float(input('Ingrese el limite superior para el valor de x: '))

#Uso de la funcion cuadratica
x=np.linspace(x_1,x_2, num=resolucion) 
y=f(a,b,c,x)

#Este codigo permite dibujar 
fig, ax=plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0,color='r')     #para crear una linea horizontal
ax.axvline(x=0,color='r')     #para crear una linea vertical
```

# **FUNCIÓN CUADRÁTICA EN FORMA CANÓNICA**

---

La función cuadrática en su forma canónica es: 
> $y=f(x)=a(x-h)^{2} + k$

Donde:
*   **h** es el valor en **x** del vértice de la párabola(figura que se construye de una función cuadrática).
*   **k** es el valor en **y** del vértice de la parábola.
*   **a** es el coeficiente que permite modificar el trazo de la parábola.


Además es importante aclarar que el dominio de esta función son todos los números reales ℛ, osea no existe restricción alguna para los valores de x (la variable independiente).

```python
def f(a,h,k,x): 
  return a*(x-h)**2+k

#Creacion de valores para graficar
a=float(input('Ingrese el valor del coeficiente a: '))
h=float(input('Ingrese el valor del coeficiente h: '))
k=float(input('Ingrese el valor del coeficiente k: '))

#Creacion de puntos y limites para la variable independiente x
resolucion=int(input('Ingrese el numero de puntos: '))
x_1=float(input('Ingrese el limite inferior para el valor de x: '))
x_2=float(input('Ingrese el limite superior para el valor de x: '))

#Uso de la funcion cuadratica
x=np.linspace(x_1,x_2, num=resolucion) 
y=f(a,h,k,x)

#Si deseo observar solo los puntos: print(x,y)

#Este codigo permite dibujar 
fig, ax=plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0,color='r')     #para crear una linea horizontal
ax.axvline(x=0,color='r')     #para crear una linea vertical

```
